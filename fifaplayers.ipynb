```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Convert CSV File to DataFrame
dataframe = pd.read_csv("fifaPlayers.csv")

#Set the query to display all the columns
pd.options.display.max_columns = len(dataframe.columns)

#Print 5 records from the dataframe. head() prints 5 by default
dataframe.head()

```

```python
#Display all the coulmn names
dataframe.columns
```

```python
#Covert the columns name to list to be able to view each coulmn name
dataframe.columns.tolist()
```

```python
#Number of rows and column (rows, column)
dataframe.shape
```

```python
#Count the number of player in each country
dataframe.groupby('nationality').size()
```

```python
#Top 10 contries for the highest number of player
dataframe.groupby('nationality').size().nlargest(10)
```

```python
#Top 5 Countries for the highest number of player
topFiveCountry =  dataframe.groupby('nationality').size().nlargest(5)

# Horizontal Bar Plot
plt.style.use('ggplot')
ax = topFiveCountry.plot(kind='barh', color='#61D199', title='Top 5 Countries with Highest Number Of Players')
ax.set_xlabel('Players Count')
ax.set_ylabel('Nationality')
```

```python
#Top 5 players with highest stat and their respective wage 
topFiveWages = dataframe.sort_values('overall', ascending=False).head()
topFiveWages[['short_name', 'wage_eur']]
```

```python
#Top 5 players with highest wage
topFiveHighestPaid = dataframe.sort_values('wage_eur', ascending=False).head()
topFiveHighestPaid[['short_name', 'wage_eur']]
```

```python
#Vertical Bar Plot for player with highest salary 
plt.style.use('ggplot')
ax = topFiveHighestPaid.plot(kind='bar', x='short_name', y='wage_eur', color='#61D199', title='Top 5 Players with Highest Salary')
ax.set_xlabel('Player')
ax.set_ylabel('Salary')
```

```python
#Group players by their nationality
nationality = dataframe.groupby('nationality')

#Top 10 players from Germany based on their overall score
nationality.get_group('Germany').sort_values('overall', ascending=False).head(10)
```

```python
#Top 5 German player based on their height, weight and wage respectively
nationality.get_group('Germany').sort_values(['height_cm','weight_kg', 'wage_eur'], ascending=False).head()
```

```python
#Top 5 German players with higest wage
topFiveHighestPaidGerman = nationality.get_group('Germany').sort_values('wage_eur', ascending=False).head()
topFiveHighestPaidGerman[['short_name', 'wage_eur']]
```

```python
#Top 5 fifa players with highest shooting stats
topFiveInShooting = dataframe.sort_values('shooting', ascending=False).head()
topFiveInShooting[['short_name', 'shooting']]
```

```python
#Top 5 players records (short name, defending, nationality, and club) that have awesome defending skills
topFiveInDefending = dataframe.sort_values('defending', ascending=False).head()
topFiveInDefending[['short_name', 'defending', 'nationality', 'club']]
```

```python
#Group the dataframe based on the club
club = dataframe.groupby('club')

#Wages records of top 5 players of 'Real Madrid' team
RealMadrid = club.get_group('Real Madrid')
topWage = RealMadrid.sort_values('wage_eur', ascending=False)
topWage[['short_name', 'wage_eur']].head()
```

```python
#Shooting records of top 5 players of 'Real Madrid' team.
topShooting = RealMadrid.sort_values('shooting', ascending=False)
topShooting[['short_name', 'shooting']].head()
```

```python
#Defending records of top 5 players of 'Real Madrid' team
topDefending = RealMadrid.sort_values('defending', ascending=False)
topDefending[['short_name', 'defending']].head()
```

```python
#Nationality records of top 5 players of 'Real Madrid' team
playerNationality = RealMadrid.sort_values('overall', ascending=False)
playerNationality[['short_name', 'overall', 'nationality']].head()
```

```python
#Get the best club based on mean of an overall of score
club.agg({'overall': 'mean',}).sort_values('overall', ascending=False).head(10)
```
